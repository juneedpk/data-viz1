```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('planets')

df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)

sns.scatterplot(x="distance", y="mass",
                hue="year",palette="ch:r=-.2,d=.3_r",
                sizes=(1, 8), linewidth=0,
                data=df, ax=ax)
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")


# Plot the responses for different events and regions
sns.lineplot(x="distance", y="mass",
             hue="year",
             data=df)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark")


# Draw a combo histogram and scatterplot with density contours
f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(data=df,x='distance', y='mass', s=5, color=".15")
sns.histplot(data=df,x='distance', y='mass', bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(data=df,x='distance', y='mass', levels=5, color="w", linewidths=1)
```
