```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")

f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)

sns.scatterplot(x="Year", y="Value",
                hue="Area",palette="ch:r=-.2,d=.3_r",
                sizes=(1, 8), linewidth=0,
                data=df, ax=ax)
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")


# Plot the responses for different events and regions
sns.lineplot(x="Year", y="Value",
             data=df)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")


# Draw a combo histogram and scatterplot with density contours
f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(data=df,x='Year', y='Value', s=5, color=".15")
sns.histplot(data=df,x='Year', y='Value', bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(data=df,x='Year', y='Value', levels=5, color="b", linewidths=1)
```
